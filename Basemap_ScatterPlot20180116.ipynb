```python
#%matplotlib inline 
#plt.figure(figsize=(10,8),dpi=100)

for j in range(268):   # 1911-1644+1=268
    plt.figure(figsize=(10,8),dpi=100)
    m = Basemap(llcrnrlon=77, llcrnrlat=14, urcrnrlon=140, urcrnrlat=51, projection='lcc', lat_1=33, lat_2=45, lon_0=100)
    m.readshapefile(r'C:\Users\Wang\Desktop\PythonTest\CHN_adm_shp\CHN_adm1', 'states', drawbounds=True,color='white')
    # 绘制经纬线
    m.drawparallels(np.arange(-90., 91., 10.), labels=[1,1,0,0], fontsize=10)
    m.drawmeridians(np.arange(-180., 181., 20.), labels=[0,0,0,1], fontsize=10)

    m.fillcontinents('black', 
                 lake_color='white')
    year = j + 1644
    j= j + 3    
    #scatter plot
    
    x1 = []
    y1 = []
    x2 = []
    y2 = []
    x3 = []
    y3 = []
    x4 = []
    y4 = []
    numbers1_5 = []
    numbers6_10 = []
    numbers11_15 = []
    numbers16 = []
    for i in range(len(df)):
        numbers = df.iloc[i,j]
        lons = float(df.iloc[i,0])
        lats = float(df.iloc[i,1])
        if numbers >= 1 and numbers <= 5:
            x1.append(lons)
            y1.append(lats)
            numbers1_5.append(numbers)
        elif numbers >= 6 and numbers <= 10: 
            x2.append(lons)
            y2.append(lats)
            numbers6_10.append(numbers)
        elif numbers >= 11 and numbers <= 15: 
            x3.append(lons)
            y3.append(lats)
            numbers11_15.append(numbers)        
        elif numbers >= 16: 
            x4.append(lons)
            y4.append(lats)
            numbers16.append(numbers)    
        
    x, y = m(x1, y1)   # Convert coords to projected place in figure          
    m.scatter(x, y,  
              s=np.array(numbers1_5)*30, 
              marker="o", 
              color='#32a9f6',
              zorder=10,
              label='IDs 1-5',
              alpha=0.8)  
    #if not x2:
        #print('Hello')
    x, y = m(x2, y2)
    m.scatter(x, y,  
              s=np.array(numbers6_10)*30, 
              marker="o", 
              color='#32caf6',
              zorder=10,
              label='IDs 6-10',
              alpha=0.8)
    #if not x3:
    x, y = m(x3, y3)
    m.scatter(x, y,  
              s=np.array(numbers11_15)*30, 
              marker="o", 
              color='#32caf6',
              zorder=10,
              label='IDs 11-15',
              alpha=0.8)
    #if not x4:
    x, y = m(x4, y4)
    m.scatter(x, y,  
              s=np.array(numbers16)*30, 
              marker="o", 
              color='#32caf6',
              zorder=10,
              label='IDs 16-',
              alpha=0.8)    
    
    plt.legend(loc=1)
    plt.title('Precipitation 10 53 '+str(year)+' IDs')
    plt.savefig(u'K:\\20171027資料前處理\\變數提取\\降水10和降水異類53\\'+'Precipitation10and53_'+str(year)+'_IDs.jpg')
    plt.close()
    #plt.show()
```
